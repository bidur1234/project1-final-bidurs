```python
# Your code here - remember to use markdown cells for comments as well!
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
%matplotlib inline
```

```python
df1=pd.read_csv('zippedData/bom.movie_gross.csv.gz')
```

```python
df1.head()
```

```python
#Display info of df1
df1.info()
```

```python
df2=pd.read_csv('zippedData/imdb.title.basics.csv.gz')
```

```python
df2.head()
```

```python
#Display info of df2
df2.info()
```

```python
df3=pd.read_csv('zippedData/imdb.title.ratings.csv.gz')
```

```python
df3.head()
```

```python
#Display info of df3
df3.info()
```

```python
#Data Cleaning
#convert df1 'domestic_gross' from str to int.
df1['domestic_gross'] = pd.to_numeric(df1['domestic_gross'], errors='coerce').astype('Int64')
df1.info()
```

```python
#convert df1'foreign_gross' from str to int.
df1['foreign_gross'] = pd.to_numeric(df1['foreign_gross'], errors='coerce').astype('Int64')
df1.info()
```

```python
print(df1)
```

```python
# adding up new column worldwide_gross in df1
df1['worldwide_gross']= df1['domestic_gross']+df1['foreign_gross']
print(df1)
```

```python
# After adding up new column in df1
df1.info()
```

```python
print(df1.head())
print(df2.head())
print(df3.head())
```

```python
# droping unused columns in df1
df1=df1.drop(['domestic_gross', 'foreign_gross'], axis=1)
```

```python
# displaying after droping columns from df1
df1.head()
```

```python
# droping unused columns in df2
df2=df2.drop(['original_title', 'start_year'], axis=1)
```

```python
# displaying after droping columns from df2
df2.head()
```

```python
#Data cleaning to eliminate duplicate row
#eliminate any duplicates in df1 and reassign.
df1 = df1.drop_duplicates(subset=None, keep='first')
display(df1)
#eliminate any duplicates in df2 and reassign.
df2 = df2.drop_duplicates(subset=None, keep='first')
display(df2)
#eliminate any duplicates in df3 and reassign.
df3 = df3.drop_duplicates(subset=None, keep='first')
display(df3)
```

```python
#Data cleaning to eliminate NaN values
# check to see how many "NaN" values in df1
display(df1.isna().sum())
#drop the NAN rows
df1=df1.dropna()
```

```python
#Data cleaning to eliminate NaN values
# check to see how many "NaN" values in df2
display(df2.isna().sum())
#drop the NAN rows
df2=df2.dropna()
```

```python
#Data cleaning to eliminate NaN values
# check to see how many "NaN" values in df3
display(df3.isna().sum())
#No drop NAN needed becuae NaN none
```

```python
#We will change "title" in df1 and "primary_title" in df2 to "movie" to match, and make the "movie" column name consistent across the 3 data sets
df1=df1.rename(columns={"title": "movie"})
df2= df2.rename(columns = {"primary_title": "movie"})
display(df1.head())
display(df2.head())


```

```python
#Merging Datasets
merged_df4=pd.merge(df1,df2, on='movie', how='inner')
```

```python
merged_df4.head()
```

```python
#merging merged_df4 and df3
merged_df5=pd.merge(merged_df4, df3, on='tconst', how='inner')
```

```python
merged_df5.head()
```

```python
#sort dataframe by movies alphabetically
merged_df5 = merged_df5.sort_values(by='movie', ascending=True)
merged_df5.head()
```

```python
#Post-merge cleanup
#run groupby and take only the first instance of each movie group
df6= merged_df5.groupby('movie')
df6.first()
#reassign back to a dataframe object
merged_movie_df6= df6.first()
merged_movie_df6
```

```python
# Manipulating "genres" data
#movies having more than one genre
#convertinf all data for movies with more than one genre
#for analysis purposes, change all multiple genre observations into the string "mix"
merged_movie_df6.loc[merged_movie_df6['genres'].str.contains(','), 'genres'] = 'mix'
merged_movie_df6.head()
```

```python
#final_dataframe
final_df=merged_movie_df6.copy()
```

```python
final_df.head()
```

```python
#Data Visualization

#year vs. average_revenue
#studio vs. average_revenue
#genres vs. average_revenue
#ratings vs. average_revenue



```

```python
#Getting the average revenue per year
year_average = final_df[['year', 'worldwide_gross']]
year_average_grouped = year_average.groupby('year', as_index=False).mean()
year_average_grouped
```

```python
fig, ax = plt.subplots()
fig.set_size_inches(20,10)

ax = sns.lineplot(x="year", y="worldwide_gross", marker="o", ci=0, markersize=15, data=year_average_grouped)
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter(1))
plt.yticks(ax.get_yticks(), ax.get_yticks()/1000000)
plt.ylabel('Average Movie Gross Revenue (In Millions Of U.S. Dollars)', fontsize=15)
plt.xlabel('Movie Release Year', fontsize=15)
sns.set(style='whitegrid', font_scale=1.2)
plt.show()
```

```python
#generes vs.average_revenue
genres_average = final_df[['genres', 'worldwide_gross']]
genres_average_grouped = genres_average.groupby('genres', as_index=False).mean()
genres_average_grouped_sorted = genres_average_grouped.sort_values(by='worldwide_gross', ascending=True)
genres_average_grouped_sorted
```

```python
genres_average_grouped_sorted = genres_average_grouped_sorted.drop(genres_average_grouped_sorted.index[:7])
genres_average_grouped_sorted
```

```python
#scatterplot of genres vs. worldwide_gross (average)
sns.set(font_scale = 5)
sns.set(style="whitegrid", color_codes=True)
pal = sns.color_palette("Greens_d", len(genres_average_grouped_sorted))
fig, ax = plt.subplots()
fig.set_size_inches(15,10)
ax = sns.barplot(x="genres", y="worldwide_gross", ci=None, palette=np.array(pal[::-1]),data=genres_average_grouped_sorted)
#ax.set(xlabel='Movie Genre', ylabel='Average Movie Gross Revenue (U.S. Dollars)')
plt.xlabel('Movie Genres', fontsize=25, weight = 'bold')
plt.ylabel('Average Movie Gross Revenue (In Millions Of U.S. Dollars)', fontsize=25, weight = 'bold')
plt.yticks(ax.get_yticks(), ax.get_yticks()/1000000)
plt.tick_params(axis='both', which='major', labelsize=20)
#sns.set_style('darkgrid')
plt.show()

#Observation
#Sci-fi genre movies were the highest earner at the box office. 
```

```python
#studio vs.average_revenue
studio_average = final_df[['studio', 'worldwide_gross']]
studio_average_grouped = studio_average.groupby('studio', as_index=False).mean().head(10)
studio_average_grouped
```

```python
# Top 10 studio in terms of Revenue generating at the global box office
studio_average_grouped_top10 = studio_average_grouped.sort_values(by= 'worldwide_gross', ascending=False).head(10)
studio_average_grouped_top10
```

```python
#scatterplot of studio vs. worldwide_gross (average)
sns.set(font_scale = 5)
sns.set(style="whitegrid", color_codes=True)
pal = sns.color_palette("Greens_d", len(studio_average_grouped_top10))
fig, ax = plt.subplots()
fig.set_size_inches(20,10)
ax = sns.barplot(x="studio", y="worldwide_gross", ci=None, palette=np.array(pal[::-1]),data=studio_average_grouped_top10)
#ax.set(xlabel='Studio', ylabel='Average Movie Gross Revenue (U.S. Dollars)')
plt.xlabel('Movie Studio', fontsize=25, weight = 'bold')
plt.ylabel('Average Movie Gross Revenue (In Millions Of U.S. Dollars)', fontsize=25, weight = 'bold')
plt.yticks(ax.get_yticks(), ax.get_yticks()/1000000)
plt.tick_params(axis='both', which='major', labelsize=20)
#sns.set_style('darkgrid')
plt.show()
```

```python
#Ensuring numeric data types
final_df['averagerating']=pd.to_numeric(final_df['averagerating'], errors='coerce')
final_df.info()
```

```python
#Averageratings vs. Worldwide_gross
# Correlation between Averagerating and worldwide_gross
correlation= final_df['averagerating'].astype(float).corr(final_df['worldwide_gross'].astype(float))
print(f'Correlation between Ratings and Gross: {correlation}')

#Observation
#The result shows that there is a positive correlation between Averagerating and worldwid_gross.
#It means as the ratings of movies increase, their gross earnings tend to increase.
#But the correlation is not very strong
```

```python
#scatterplot of averagerating vs. worldwide_gross
fig, ax = plt.subplots()
fig.set_size_inches(18,12)

ax = sns.scatterplot(x="averagerating", y="worldwide_gross", marker="o", ci=68, data=final_df)
ax.xaxis.set_major_locator(ticker.MultipleLocator(10))
ax.xaxis.set_major_formatter(ticker.ScalarFormatter())
ax.set(xlabel='Movie Rating', ylabel='Worldwide Movie Gross Revenue (U.S. Dollars)')
sns.set(style='darkgrid', font_scale=1.5)
plt.show()
```
